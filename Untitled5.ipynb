```python

%matplotlib notebook
import matplotlib.pyplot
import numpy as np
from matplotlib import animation
import matplotlib.colors
cmap = matplotlib.colors.ListedColormap(["#0099ff", "#d5d5c3", "#004d00", "#008000", "#00b300", "#4dff4d", "#ff8533",
                                         "#787854"])
np.random.seed(0)

#NORD : DIRECTION
#SUD : DIRECTION
#EST : DIRECTION
#OUEST : DIRECTION
def hasard(p):
    r = np.random.random()
    assert 0 <= p <= 1
    return r <= p


def creerForet(n, m, pcoccup, pw):
    """creer une foret de taille n,m avec pcoccup la proportion d'arbe et pw la proportion d'eau sur les elements incombustible"""
    foret = np.zeros((n, m))
    for i in range(n):
        for j in range(m):
            if hasard(pcoccup):
                k = np.random.random()
                if k < 0.25:
                    foret[i, j] = 1.
                elif k < 0.5:
                    foret[i, j] = 2.
                elif k < 0.75:
                    foret[i, j] = 3.
                else:
                    foret[i, j] = 4.
            elif np.random.random() < pw:
                foret[i, j] = -1.
    return foret

def burnprob(foret, i, j):
    """defini le nombre de case d'eau a proximiter d'une case de vegetation """
    n, m = foret.shape
    u = 0
    if 1. <= foret[i, j] <= 4.:
        for r in range(max(0, i-1), min(n, i+2)):
            for g in range(max(0, j-1), min(m, j+2)):
                if foret[r, g] == -1:
                    u += 0.1
    return u

def checkburn(foret, i, j):
    """verifie si une case de vegetation est susceptible de bruler"""
    n, m = foret.shape

    for y in range(max(0, i-1), min(n, i+1)):
        if foret[y, j] == 5.:
            return True
    for x in range(max(0, j - 1), min(m, j + 2)):
        if foret[i, x] == 5.:
            return True
    return False

def mapburn(foret,direction):
    """creer une carte de la probabilite que chaque case a de bruler"""
    n, m = foret.shape
    mb = np.zeros((n, m))
    for i in range(n):
        for j in range(m):
            if 1. <= foret[i, j] <= 4.:
                mb[i, j] = 1.
                mb[i, j] = 1 - burnprob(foret, i, j)
                if checkburn(foret, i, j) and not(peutbrulervent_nord(foret, i, j, direction)):
                    mb[i, j] = mb[i, j] - 0.15
    return mb


def mettreLeFeu(foret):
    """met le feu a une foret en un point"""
    n, m = foret.shape
    i = np.random.randint(0, n + 1)
    j = np.random.randint(0, m + 1)
    while foret[i, j] != 5.:
        if 1. <= foret[i, j] <= 4.:
            foret[i, j] = 5.
        else:
            i = np.random.randint(0, n)
            j = np.random.randint(0, m)
    return foret


def peutbrulervent_nord(foret, i, j, direction):
    """verifie si une case peut bruler a cause du vent"""
    n, m = foret.shape
    if direction == 'NORD':
        if i == n-1:
            return False
        elif 1. <= foret[i, j] <= 4.:
            for v in range(max(0, j-1), min(m, j+2)):
                if foret[i+1, v] == 5.:
                    return True
        else:
            return False

    elif direction == 'SUD':
        if i == 0.:
            return False
        elif 1. <= foret[i, j] <= 4.:
            for v in range(max(0, j-1), min(m, j+2)):
                if foret[i-1, v] == 5.:
                    return True
        else:
            return False
    
    
    elif direction == 'EST':
        if j == 0.:
            return False
        elif 1. <= foret[i, j] <= 4.:
            for v in range(max(0, i-1), min(m, i+2)):
                if foret[v, j-1] == 5.:
                    return True
        else:
            return False
    
    
    elif direction == 'OUEST':
        if j == m-1:
            return False
        elif 1. <= foret[i, j] <= 4.:
            for v in range(max(0, i-1), min(m, i+2)):
                if foret[v, j-1] == 5.:
                    return True
        else:
            return False

    return False

def propageFeu(foret, direction):
    """propage le feu dans la foret"""
    n, m = foret.shape  
    c = np.copy(foret)
    d = mapburn(c, direction)
    p = np.random.random()
    for i in range(n):
        for j in range(m):
            if peutbrulervent_nord(c, i, j, direction) and p < d[i, j]:
                foret[i, j] = 5.
                
            elif checkburn(c, i, j) and p < d[i, j]:
                foret[i, j] = 5.
    for i in range(n):
        for j in range(m):
            if c[i, j] == 5. and np.random.random() < p:
                foret[i, j] = 6.
    return foret

def auFeu(foret, direction):
    """verifie si il existe au moins une case qui peut bruler et s'il reste une case en feu"""
    n, m = foret.shape
    for i in range(n):
        for j in range(m):
            if peutbrulervent_nord(foret, i, j, direction) or checkburn(foret, i, j) or foret[i, j] == 5.:
                return True
    return False

def metFeuForet(foret,cmap=None):
   
    foret = mettreLeFeu(foret)
    while auFeu(foret):
        foret = propageFeu(foret, direction)
    return foret

def animationFeu(foret, direction, cmap=cmap):
    """creer l'animation de la propagation du feu dans la foret"""
    fig = matplotlib.pyplot.figure()
    film = []

    foretFeu = mettreLeFeu(foret)
    film.append([matplotlib.pyplot.matshow(foret, fignum=False, cmap=cmap, animated=True)])
    
    while auFeu(foret, direction):
        foretFeu = propageFeu(foretFeu, direction)
        film.append([matplotlib.pyplot.matshow(foretFeu, fignum=False, cmap=cmap, animated=True)])
        

    ani = animation.ArtistAnimation(fig, film, interval=1000, blit=True, repeat_delay=100)
    
    return ani
```
